```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Task 1

X_partition = 100
Y_partition = 100
left = -np.pi
right = np.pi
a = 2
b = 3
X = np.linspace(left,right,X_partition)
Y = np.linspace(left,right,Y_partition)

z = np.outer(np.sin(a*X), np.cos(b*Y))
u1 = z + np.random.normal(0,z.max() * 0.01, size=z.shape)
u2 = z + np.random.poisson(lam=z.max() * 0.01, size=z.shape)
dif_1 = z - u1
dif_2 = z - u2

data_list = [z, u1, dif_1,
             z, u2, dif_2]

ticks_X = np.linspace(0, X_partition, 5)
ticks_Y = np.linspace(0, Y_partition, 5)
tick_labels = [r"$-\pi$", r"$-\pi/2$", "0", r"$\pi/2$", r"$\pi$"]

fig, axes = plt.subplots(2, 3, figsize=(18, 10), gridspec_kw={'width_ratios': [1, 1, 1.5]})
for i, ax in enumerate(axes.flat):
    im = ax.imshow(data_list[i], cmap="coolwarm", vmin=data_list[i].min(), vmax=data_list[i].max())
    ax.set_title(f'{chr(97 + i)})')
    ax.set_xticks(ticks_X)
    ax.set_xticklabels(tick_labels)
    ax.set_yticks(ticks_Y)
    ax.set_yticklabels(tick_labels)
    if(i == 4):
        cbar_ax = fig.add_axes([0.58, 0.1, 0.015, 0.8])
        fig.colorbar(im, cax=cbar_ax)


    if(i%3 == 2):
        fig.colorbar(im, ax=ax) # colorbar   


plt.show()
```
