```python
import matplotlib.pyplot as plt
import numpy as np

# Task 1

X_partition = 20
Y_partition = 20
left = -np.pi
right = np.pi
a = 2
b = 3
X = np.linspace(left,right,X_partition)
Y = np.linspace(left,right,Y_partition)

z = np.outer(np.sin(a*X), np.cos(b*Y))
u1 = z + np.random.normal(0,z.max() * 0.05)
u2 = z + np.random.poisson(3)

data_list = [z, u1, z - u1,
             z, u2, z - u2]
fig, axes = plt.subplots(2, 3, figsize=(15, 10))


ticks_X = np.linspace(0, X_partition, 5)
ticks_Y = np.linspace(0, Y_partition, 5)
tick_labels = [r"$-\pi$", r"$-\pi/2$", "0", r"$\pi/2$", r"$\pi$"]

title = 'a'
for i, ax in enumerate(axes.flat):
    im = ax.imshow(data_list[i], cmap="coolwarm")
    ax.set_title(f'{i}')
    fig.colorbar(im, ax=ax)
    ax.set_xticks(ticks_X)
    ax.set_xticklabels(tick_labels)
    ax.set_yticks(ticks_Y)
    ax.set_yticklabels(tick_labels)

plt.tight_layout()
plt.show()
```
